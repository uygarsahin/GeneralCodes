```python
import base64 

def b64e(s): 
    return base64.b64encode(s.encode()).decode() 
def b64d(s): 
    return base64.b64decode(s).decode()
```

```python
#Encryption
b64e('1huEmwQlPbaMFXAU')
```

```python
#Decryption
base64.b64decode('YXNfanp3eGZnQG5hbS5jb3JwLmdtLmNvbQ==').decode()
```

```python
import base64 

#Encryption of String
def b64e(string): 
    return base64.b64encode(string.encode()).decode() 
#Decryption of String 
def b64d(string): 
    return base64.b64decode(string).decode()

WordOfChange ='Hello World'

OutPutofEncWord=b64e(WordOfChange)

print(OutPutofEncWord)
print('='*10)

OutPutofDencWord=b64d(OutPutofEncWord)

print(OutPutofDencWord)
```
